```python
path = "../input"

files = list_files_from_dir(path)
data_dict = {}
new_dict_list = []
for i in range(30,59):
    new_dict_list.append(f"iteration:{i}")
for file in files:
    print(file)
    new_dict = load_json(path + "/" + file)
    if "batch2" in file:
        for test_per in new_dict:
            final_dict = {}
            final_dict[test_per] = dict(zip(new_dict_list, list(new_dict[test_per].values())))
            if test_per in data_dict:
                data_dict = {**data_dict[test_per], **final_dict[test_per]}
            else:
                data_dict[test_per] = {}
                data_dict = {**data_dict[test_per], **final_dict[test_per]}
    else:
        data_dict = {**data_dict, **new_dict}
```

```python

data_list = []
F1_dict = {}
for test_per in data_dict:
    for replica in data_dict[test_per]:
        print(replica)
        for model in data_dict[test_per][replica]:
            for epoch, value in data_dict[test_per][replica][model].items():
                for f1 in value:
                    if f"{test_per}_{model}_{epoch}" in F1_dict:
                        F1_dict[f"{test_per}_{model}_{epoch}"].append(f1["avg_f1"][1])
                    else:
                        F1_dict[f"{test_per}_{model}_{epoch}"] = []
                        F1_dict[f"{test_per}_{model}_{epoch}"].append(f1["avg_f1"][1])
                    
#                     data_list.append([test_per, model, epoch, f1["avg_f1"][0]])
        
```
