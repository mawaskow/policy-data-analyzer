```python
import pandas as pd
import json
import sys
# sys.path.append("C:/Users/jordi/Documents/GitHub/policy-data-analyzer/tasks/")
sys.path.append("/home/propietari/Documents/GitHub/policy-data-analyzer/tasks/")
from data_loading.src.utils import *
```

```python
path = "../input"

files = list_files_from_dir(path)
data_dict = {}
for file in files:
    data_dict = {**data_dict, **load_json(path + "/" + file)}
```

```python

data_list = []
for test_per in data_dict:
    for model in data_dict[test_per]:
        for epoch, value in data_dict[test_per][model].items():
            for f1 in value:
                data_list.append([test_per, model, epoch, f1["avg_f1"][0]])
        
```

```python
df = pd.DataFrame (data_list,columns=['Test_percentage', 'Model', 'Epoch', 'Avg_F1'])
```

```python
df.groupby(["Test_percentage", "Epoch"]).mean()
```
