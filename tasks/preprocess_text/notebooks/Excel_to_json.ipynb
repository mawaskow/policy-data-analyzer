# Structuring excell sheet info into json

## Imports 

```python
from pathlib import Path
import json
import pandas as pd
import numpy as np
```

## Loading original data form excel

```python
path = Path("../input/")
filename = "WRI Policy Tags Complete File phrases extraction all.xlsx"
file = path / filename
df = pd.read_excel(file) 
```

```python
print(df.shape)
data = df.to_numpy()
df.head(15)
```

```python
dict = {}

i = 0
for item in data:
    if ";" in item[2]:
        sentences = item[2].split(";")
    else:
        sentences = item[2]
    for sentence in sentences:
        if sentence != "":
            sentence = sentence.lstrip().rstrip()
            i += 1
            print(i, "----", sentence)
            if item[0] in dict:
                dict[item[0]]["Seccion unica"]["sentences"][str(i)] = {"text" : sentence, "labels" : item [4]}
            else:
                dict[item[0]] = {"Seccion unica" : {"tags" : [], "sentences" : {}}}
                dict[item[0]]["Seccion unica"]["sentences"][str(i)] = {"text" : sentence, "labels" : item [4]}
    #         print(sentence, "\n***\n")
```

```python
path = Path("../output/")
filename = "tagged_sentences_all.json"
file = path / filename
with open(file, 'w') as f:
    json.dump(dict, f, indent = 4)
```
